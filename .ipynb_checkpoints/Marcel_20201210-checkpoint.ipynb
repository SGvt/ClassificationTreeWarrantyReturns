# Introduction

If the goal here is not to discuss the different ways to approach After-Sales strategy, it nonetheless seemed interesting to contextualize the business case in order to give this matter its full depth.


<img caption="Business Process" src="Illustrations/BusinessProcess.png" style="width: auto; height: 250px;" align="right" > 
<font size="3"><b>After-Sales support, a marketing tool in itself</b></font><br>
Sometimes neglected, the After-Sales Support can (should?) be an integrated part of the Company’s overall marketing strategy. The quality of a company’s After-Sales service can be decisive in the consumer buying process also referred to as consumer buying cycle. A performant after-sales support can result in an increased brand loyalty and positive word-of-mouth to name just a few.

<font size="3"><b>Warranty returns and subsequent actions</b></font><br>
One aspect of the after-sale support is dealing with warranty returns. Depending on the industry sector and company’s warranty policy, those returns can have different issues. Replacement, repair or buying a new product at a discounted rate are among the most common solutions offered to customers. Choosing between these issues can be a complicated and tedious process depending on the type of product and scenarios.







<font size="3"><b>A decision-making process prone to variations</b></font><br>
Usually a representative is responsible for selecting the most appropriate offer .and involves different criteria. Some are intrinsic to the case presented (i.e. features of the product, context, …) some are tied to the individual who’s processing the case (i.e. subjectivity, level of awareness, …). The later introduces a form of bias that can result in important variations in the way cases are handled by agents.

<img caption="Business Process" src="Illustrations/WarrantyReturnsScenario.png" style="width: auto; height: 300px;" align="center" > 

<font size="3"><b>Bias reduction</b></font><br>
Training and experience are two parameters that tend to reduce this bias. Training easy to manage for a company and if well designed can also steepen representatives’ learning curve and having them reaching their optimum efficiency earlier providing that they stay in the role long enough…

<img caption="Business Process" src="Illustrations/LearningCurveCircle.png" style="width: auto; height: 400px;" align="right" > 

<font size="3"><b>Turnover rate the snag</b></font><br> 
Unfortunately, another parameter comes into play and throws sands into the machine: turnover rate. Customer Service Representative roles are linked to a very high turnover rate, despite all the efforts made by companies to limit it, it seems that this is inherent to the role and this issue can grow exponentially depending on how big the company is. 

According to the 2016 US Contact Center Decision Makers’ Guide (ContactBabel.com) the Average Annual Turnover Rate for a Customer Service Representative (CSR) was in average 27% in North America. This number varies by industry sector and ranges from a low 21% in the Financial Service Sector to a higher 53% for the Third-Party Teleservices Providers.

A question now arises: <b> “What can be done to unify as much as possible the company’s voice providing the bias introduced by individuals and high turnover rate?”</b>





<img caption="Grey Area" src="Illustrations/customerfocus.jpg" style="width: auto; height: 350px;" align="left"> 

<font size="3"><b>Decision Tree to the rescue</b></font><br>
A potential avenue would be assisting the decision-making process. A multitude of tools are available, here we are going to focus on one: Decision Trees. One big advantage of Decision Trees is their “transparency”. Decision Trees are sometimes referred to as White-Box models because it is easy to explain how they behave. You can learn more about decision trees here.



<font size="3"><b>Re-focusing on Customer Experience</b></font><br>
On top of minimizing the impact of time/experience on the Company’s Voice, assisting the decision-making process would also reduce the stress factor linked to the action of making a decision and improve the overall efficiency and robustness of the workflow. As a result, customer service representatives will be able to refocus on communication and ultimately contribute to reinforcing customer relationships by elevating their experience with the Brand.

# Decision Tree

<img caption="Picture" src="Illustrations/forest.jpg" style="width: auto; height: 300px;" align="right"> 
Various decision tree algorithms are available, Iterative Dichotomiser 3 (ID3), C4.5 which is the successor of ID3, C5.0 an improved version of C4.5 or the Classification and Regression Trees (CART). The latest is similar to C4.5 but also has the capability to deal with regression problems (i.e. numerical target variables).

The method used here after (Scikit-learn library) is an optimised version of the CART algorithm. 

This algorithm produces a binary tree (i.e. each node splits the input space into 2 subspaces) using a greedy approach. Each node is defined as the feature that would minimize the cost function. For classification problems, the cost function is usually associated to two criteria: Gini impurity or Entropy.

## Pre-Processing

The features composing our dataset are either categorical (nominal), binary or discrete variables.
Unfortunately the CART method does not accept categorical variables as inputs.

To get around this issue, a popular technique is going to be used: One-Hot-Encoding (OHE). 
For each feature, OHE translates each unique value into a binary vector.

<img caption="One Hot Encoding" src="Illustrations/OneHotEncoding.png" style="width: auto; height: 100px;" align="right" >

Let's illustrate the effect of one-hot encoding predictors using  one of our predictor, "Fabric" here. The "Fabric" feature originaly contains four unique values [Down, Gore, Paclite, Other]. One-hot encoding it results in four new variables, one for each unique value. These new variables are called binary variables because they only contain 0 and 1.

Note: For coherence sake, it would have made sense to also encode the target variable but useless as CART algorithm is able to handle target variables of a categorical type (Label Encoding method could have been used here as the target variable can only be in the form of a one dimensional vector).

As mentioned earlier, some of our criteria materialize defects found on products. These fields hold unstructured data describing the location of the issue. Because they are not normalized, we're going to simplify the approach by bynary encoding them ("Yes/No" if the issue has been flagged). This choice reduces the depth of the problem but is deemed acceptable for a first approach.

```python
# Definition of the feature to be used for the decision tree
featuresdt=['Repair Threshold',  
            'Fabric_Down Loft', 'Fabric_K Fleece', 'Fabric_WBW 2L Membrane', 'Fabric_WBW 3L Membrane', 
            'Fabric failure - Membrane delamination_No', 'Fabric failure - Membrane delamination_Yes', 
            'Fabric failure - Leak through_No', 'Fabric failure - Leak through_Yes', 
            'Fabric failure - Wetting out_No', 'Fabric failure - Wetting out_Yes', 
            'Fabric failure - Down leakage_No', 'Fabric failure - Down leakage_Yes', 
            'Fabric failure - Fading_No', 'Fabric failure - Fading_Yes', 'Fabric failure - Pilling_No', 
            'Fabric failure - Pilling_Yes', 'Fabric torn - Rips - Patchable_No', 'Fabric torn - Rips - Patchable_Yes', 
            'Fabric torn - Rips - Not patchable - 1 panel_No', 'Fabric torn - Rips - Not patchable - 1 panel_Yes', 
            'Fabric torn - Rips - Not patchable - X panels_No', 'Fabric torn - Rips - Not patchable - X panels_Yes', 
            'Fabric torn - Abrasion - Light_No', 'Fabric torn - Abrasion - Light_Yes', 'Fabric torn - Abrasion - Heavy - 1 panel_No', 
            'Fabric torn - Abrasion - Heavy - 1 panel_Yes', 'Fabric torn - Abrasion - Heavy - X panels_No', 
            'Fabric torn - Abrasion - Heavy - X panels_Yes', 'Fabric torn - Burn - Patchable - _No', 
            'Fabric torn - Burn - Not patchable - 1 panel_No', 'Fabric torn - Burn - Not patchable - 1 panel_Yes', 
            'Fabric torn - Burn - Not patchable - X panels_No', 'Fabric torn - Burn - Not patchable - X panels_Yes', 
            'Fabric torn - Stain_No', 'Fabric torn - Stain_Yes', 'Fabric torn - Membrane contamination - Light_No', 
            'Fabric torn - Membrane contamination - Light_Yes', 'Fabric torn - Membrane contamination - Heavy_No', 
            'Fabric torn - Membrane contamination - Heavy_Yes', 'Zipper failure - Zipper structureMain_No', 
            'Zipper failure - Zipper structureMain_Yes', 'Zipper failure - Zipper structureAll except main_No', 
            'Zipper failure - Zipper structureAll except main_Yes', 'Zipper failure - Zipper slider_No', 
            'Zipper failure - Zipper slider_Yes', 'Delamination - Press_No', 'Delamination - Press_Yes', 
            'Delamination - 70D Nylon_No', 'Delamination - 70D Nylon_Yes', 'Delamination - Seam tape_No', 
            'Delamination - Seam tape_Yes', 'Seam failure - Seam sewn non-WBW Membrane_No', 
            'Seam failure - Seam sewn non-WBW Membrane_Yes', 'Delamination - Seam sewn WBW Membrane_No', 
            'Delamination - Seam sewn WBW Membrane_Yes', 'Delamination - Seam laminated_No', 
            'Other - Buckle_No', 'Other - Buckle_Yes', 'Known_issue_No', 'Known_issue_Yes', 
            'Repair solution_No', 'Repair solution_Yes', 'Repair solution_Yes ', 'Age_0', 'Age_1', 'Age_2', 
            'Age_3', 'Age_4', 'Age_5', 'Age_6', 'Age_7', 'Age_8', 'Age_9', 'Age_10', 'Age_11', 'Age_12', 
            'Age_13', 'Age_14', 'Age_15', 'Age_16', 'Age_17', 'Age_18', 'Age_19', 'Age_20', 
            'Repair Threshold Status_above', 'Repair Threshold Status_below', 'Repair Threshold Status_equal']

len(featuresdt)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

# Predictor and response variables definition
X = dfoh[featuresdt] # Predictor variables
Y = dfoh['Customer Offer'] # Target variable

# Split Data Training/Testing
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state=0, stratify=Y) 

# Create Decision Tree
clf = DecisionTreeClassifier(criterion = "gini", random_state = 0, max_depth=3)

# Train Model
clf = clf.fit(X_train, y_train)

accuracy = clf.score(X_test, y_test)

print("Training sample size is {}, testing sample size is {} and the Decision tree accuracy is {:.2f}%  ".format(len(X_train), len(X_test), accuracy*100))
```

<b>[Observations]</b><br>
- <b>Training set size</b>: 75% of the dataset has been used to train the model,
- <b>Depth</b>: For readability sake above model has been limited to 3 nodes (i.e. tree's max depth 3),
- <b>Criterion</b>: Classification trees define the best split at each node using a measure called impurity. 3 different criteria can be used to calculate the impurity:  Gini index, Entropy or Misclassification. Sklearn default criteria, the Gini index has been used here. These criteria usually yield very similar results,
- <b>Accuracy</b>: With the above configuration and sample, classification tree's accuracy (i.e. mean accuracy) is 71.43%. This result is already descent but can be drastically improved by tuning some of the above listed parameters and/or working on the initial dataset.

## Optimization

Many different ways to optimize decision tree. 2 main levers to pull are the features themselves, the number, their quality, the size of the set, and then levers onthe way the decision tree is cerated

 looking at all of them deserve its own focus too long but just looking at two parameters that are the depth and the training set size to see of they affect accuracy .
 
here all the features are categorical variables or discrete which in this case will behave similarly.

limited by the nature of the features all categorical or discrete




```python
# Decision Tree Optimisation Exploration Function
def dtopti(testsize, depth):
    
    """Rerun decision tree with different depth and save accuracy in a list"""
    # Split Data Training/Testing
    X_trainopti, X_testopti, y_trainopti, y_testopti = train_test_split(X, Y, test_size = testsize, random_state=0, stratify=Y)
    clfopti = DecisionTreeClassifier(criterion = "gini", random_state = 0, max_depth=depth) # Create Decision Tree
    clfopti = clfopti.fit(X_trainopti, y_trainopti) # Train Model
    y_predopti = clfopti.predict(X_testopti) # Predict response for X_test
    accuracyopti = clfopti.score(X_testopti, y_testopti)
    
    return accuracyopti

testsplit = []

for i in range(1,6,1):
    accuracy = []
    for j in range(1,40):
        accuracy.append(dtopti(float(i)/10, j))
    testsplit.append(accuracy)

# print testsplit
#print len(testsplit)
```

```python
# Visualize accuracyopti
plt.figure(figsize=(20, 10))
for serie in range(len(testsplit)):
    plt.plot(testsplit[serie],label = 'Training Set Size %s%%'%((1-float(serie+1)/10)*100)) 
    
    
plt.legend()
plt.title('Impact of Training Set Size and Depth on Accuracy')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy (%)')
plt.show() 
```
