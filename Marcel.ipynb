# Classification Tree

<img caption="Picture" src="Illustrations/forest.jpg" style="width: auto; height: 200px;" align="right"> 
Various classification tree algorithms are available, Iterative Dichotomiser 3 (ID3), C4.5 which is the successor of ID3, C5.0 an improved version of C4.5 or the Classification and Regression Trees (CART). The latest is similar to C4.5 but also has the capability to deal with regression problems (i.e. numerical target variables).

The method used here after (Scikit-learn library) is an optimized version of the CART algorithm. 

This algorithm produces a binary tree (i.e. each node splits the input space into 2 subspaces) using a greedy approach. Each node is defined as the feature that would minimize the cost function. For classification problems, the cost function is usually associated to two criteria: Gini impurity or Entropy.

## Pre-Processing

The first step is going to be identifying the classification tree's features, i.e. in which fields the business logic is actually embeded. This process can be done iteratively (cf. Optimization)

<b>Simplification, Binary Encoding</b><br>
As mentioned earlier, some of the fields materialize defects found on products. Their hold unstructured data describing the location of the issue. Because they are not normalized, we're going to simplify the approach by bynary encoding them ("Yes" when an issue has been flagged, "No" if not). This choice reduces the depth of the problem but is deemed acceptable as a first approach.

```python
# Binary encoding

featurestotransform = ['Fabric failure - Membrane delamination', 'Fabric failure - Leak through', 'Fabric failure - Wetting out', 
      'Fabric failure - Down leakage', 'Fabric failure - Fading', 'Fabric failure - Pilling', 'Fabric torn - Rips - Patchable', 
      'Fabric torn - Rips - Not patchable - 1 panel', 'Fabric torn - Rips - Not patchable - X panels', 'Fabric torn - Abrasion - Light', 
      'Fabric torn - Abrasion - Heavy - 1 panel', 'Fabric torn - Abrasion - Heavy - X panels', 'Fabric torn - Burn - Patchable - ', 
      'Fabric torn - Burn - Not patchable - 1 panel', 'Fabric torn - Burn - Not patchable - X panels', 'Fabric torn - Stain', 
      'Fabric torn - Membrane contamination - Light', 'Fabric torn - Membrane contamination - Heavy', 
      'Zipper failure - Zipper structureMain', 'Zipper failure - Zipper structureAll except main',
      'Zipper failure - Zipper slider', 'Delamination - Press', 'Delamination - 70D Nylon', 
      'Delamination - Seam tape', 'Seam failure - Seam sewn non-WBW Membrane', 'Delamination - Seam sewn WBW Membrane',
      'Delamination - Seam laminated', 'Other - Buckle']


def binaryencode(dataframe):
    
    dfbe = dataframe.copy()# Deep copy (dfpp) of our initial dataset (df)

    dfbe[featurestotransform] = dfbe[featurestotransform].applymap(lambda x:'Yes' if pd.notnull(x) else 'No')
    
    return(dfbe)
```

<b>One Hot Encoding</b><br>
Those features composing our dataset are either categorical (nominal), binary or discrete variables.
Unfortunately the CART method does not accept categorical variables as inputs.

To get around this issue, a popular technique is going to be used: One-Hot-Encoding (OHE). 
For each feature, OHE translates each unique value into a binary vector.

<img caption="One Hot Encoding" src="Illustrations/OneHotEncoding.png" style="width: auto; height: 80px;" align="right" >

Let's illustrate the effect of one-hot encoding predictors using  one of our predictor, "Fabric" here. The "Fabric" feature originaly contains four unique values [Down, Gore, Paclite, Other]. One-hot encoding it results in four new variables, one for each unique value. These new variables are called binary variables because they only contain 0 and 1.

Note: For coherence sake, it would have made sense to also encode the target variable but useless as CART algorithm is able to handle target variables of a categorical type (Label Encoding method could have been used here as the target variable can only be in the form of a one dimensional vector).

```python
# Fields to one hot encode
featurestooh = [ 'Fabric', 'Fabric failure - Membrane delamination', 'Fabric failure - Leak through', 
                'Fabric failure - Wetting out', 'Fabric failure - Down leakage', 
                'Fabric failure - Fading', 'Fabric failure - Pilling', 'Fabric torn - Rips - Patchable',
                'Fabric torn - Rips - Not patchable - 1 panel', 'Fabric torn - Rips - Not patchable - X panels', 
                'Fabric torn - Abrasion - Light', 'Fabric torn - Abrasion - Heavy - 1 panel', 
                'Fabric torn - Abrasion - Heavy - X panels', 'Fabric torn - Burn - Patchable - ', 
                'Fabric torn - Burn - Not patchable - 1 panel', 'Fabric torn - Burn - Not patchable - X panels', 
                'Fabric torn - Stain', 'Fabric torn - Membrane contamination - Light', 'Fabric torn - Membrane contamination - Heavy', 
                'Zipper failure - Zipper structureMain', 'Zipper failure - Zipper structureAll except main', 
                'Zipper failure - Zipper slider', 'Delamination - Press', 'Delamination - 70D Nylon', 
                'Delamination - Seam tape', 'Seam failure - Seam sewn non-WBW Membrane', 
                'Delamination - Seam sewn WBW Membrane', 'Delamination - Seam laminated', 
                'Other - Buckle', 'Known_issue', 'Repair solution', 'Repair Threshold Status', 'Product_type']

def onehotencode(dataframe):
  
    # One hot encoding
    dfoh = pd.get_dummies(dataframe,columns=featurestooh)
    
    # Export to CSV for further observations
    dfoh.to_csv("dfoh.csv")
    
    return(dfoh, featurestooh)

dfpp, x = onehotencode(dfbe)
dfpp
```

To train the model, the dataset is going to be randomly split into a training and a testing sets. Intuitively the training set will be used to train the model while the testing set will support the evaluation of trained model's performances.

```python
# Definition of the feature to be used for the classification tree
featuresdt=['Age', 'Fabric_Down Loft', 'Fabric_K Fleece', 
            'Fabric_WBW 2L Membrane', 'Fabric_WBW 3L Membrane', 'Fabric failure - Membrane delamination_No', 
            'Fabric failure - Membrane delamination_Yes', 'Fabric failure - Leak through_No', 
            'Fabric failure - Leak through_Yes', 'Fabric failure - Wetting out_No', 
            'Fabric failure - Wetting out_Yes', 'Fabric failure - Down leakage_No', 
            'Fabric failure - Down leakage_Yes', 'Fabric failure - Fading_No', 'Fabric failure - Fading_Yes', 
            'Fabric failure - Pilling_No', 'Fabric failure - Pilling_Yes', 'Fabric torn - Rips - Patchable_No', 
            'Fabric torn - Rips - Patchable_Yes', 'Fabric torn - Rips - Not patchable - 1 panel_No', 
            'Fabric torn - Rips - Not patchable - 1 panel_Yes', 'Fabric torn - Rips - Not patchable - X panels_No', 
            'Fabric torn - Rips - Not patchable - X panels_Yes', 'Fabric torn - Abrasion - Light_No', 
            'Fabric torn - Abrasion - Light_Yes', 'Fabric torn - Abrasion - Heavy - 1 panel_No', 
            'Fabric torn - Abrasion - Heavy - 1 panel_Yes', 'Fabric torn - Abrasion - Heavy - X panels_No', 
            'Fabric torn - Abrasion - Heavy - X panels_Yes', 'Fabric torn - Burn - Patchable - _No', 
            'Fabric torn - Burn - Not patchable - 1 panel_No', 'Fabric torn - Burn - Not patchable - 1 panel_Yes', 
            'Fabric torn - Burn - Not patchable - X panels_No', 'Fabric torn - Burn - Not patchable - X panels_Yes', 
            'Fabric torn - Stain_No', 'Fabric torn - Stain_Yes', 'Fabric torn - Membrane contamination - Light_No', 
            'Fabric torn - Membrane contamination - Light_Yes', 'Fabric torn - Membrane contamination - Heavy_No', 
            'Fabric torn - Membrane contamination - Heavy_Yes', 'Zipper failure - Zipper structureMain_No', 
            'Zipper failure - Zipper structureMain_Yes', 'Zipper failure - Zipper structureAll except main_No', 
            'Zipper failure - Zipper structureAll except main_Yes', 'Zipper failure - Zipper slider_No', 
            'Zipper failure - Zipper slider_Yes', 'Delamination - Press_No', 'Delamination - Press_Yes', 
            'Delamination - 70D Nylon_No', 'Delamination - 70D Nylon_Yes', 'Delamination - Seam tape_No', 
            'Delamination - Seam tape_Yes', 'Seam failure - Seam sewn non-WBW Membrane_No', 
            'Seam failure - Seam sewn non-WBW Membrane_Yes', 'Delamination - Seam sewn WBW Membrane_No', 
            'Delamination - Seam sewn WBW Membrane_Yes', 'Delamination - Seam laminated_No', 'Other - Buckle_No', 
            'Other - Buckle_Yes', 'Known_issue_No', 'Known_issue_Yes', 'Repair solution_No', 'Repair solution_Yes', 
            'Repair Threshold Status_above', 'Repair Threshold Status_below', 'Repair Threshold Status_equal', 
            'Product_type_shell']

len(featuresdt)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

# Predictor and response variables definition
X = dfpp[featuresdt] # Predictor variables
Y = dfpp['Customer Offer'] # Target variable

# Split Data Training/Testing
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state=0, stratify=Y) 

# Create Decision Tree
clf = DecisionTreeClassifier(criterion = "gini", random_state = 42, max_depth=3)

# Train Model
clf = clf.fit(X_train, y_train)

accuracy = clf.score(X_test, y_test)

print("Training sample size is {}, testing sample size is {} and the Decision tree accuracy is {:.2f}%  ".format(len(X_train), len(X_test), accuracy*100))
```

<b>[Observations]</b><br>
- <b>Training set size</b>: 75% of the dataset has been used to train the model,
- <b>Depth</b>: For readability sake above model has been limited to 3 nodes (i.e. tree's max depth 3),
- <b>Criterion</b>: Classification trees define the best split at each node using a measure called impurity. 3 different criteria can be used to calculate the impurity:  Gini index, Entropy or Misclassification. Sklearn default criteria, the Gini index has been used here. These criteria usually yield very similar results,
- <b>Accuracy</b>: With the above configuration and sample, classification tree's accuracy (i.e. mean accuracy) is 70.88%. This result is already descent but can be drastically improved by tuning some of the above listed parameters and/or working on the initial dataset.

Once trained, the classification tree algorithm is able to predict customer offers for any new cases.

As mentioned in the introduction, the output provided by the CART Algorithm can then be used by Customer Representatives to consolidate their position. 

Assisting the decision-making process results in the minimization of the impact of time/experience on the Company’s Voice but also allows CS to shift their focus on customer relationship management.

Let's see how this can be done through an example.

```python
# Classification tree creation
def decisiontree():
    df = pd.read_csv("RawData.csv", delimiter=';')
    xpreproc1 = binaryencode(df)
    xpreproc2, x  = onehotencode(xpreproc1)
    
    X = xpreproc2[featuresdt] # Predictor variables
    Y = xpreproc2['Customer Offer'] # Target variable

    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1, random_state=42, stratify=Y) 

    clf = DecisionTreeClassifier(criterion = "gini", random_state = 42, max_depth=15)
    clf = clf.fit(X_train, y_train)
    precision = round(clf.score(X_test, y_test),3)*100
    return  clf, precision

tree, precision = decisiontree()
precision
```

```python
# Customer offer prediction for new cases
def newcase_custoffer(casenumber, itemincasenumber):
    df = pd.read_csv("RawData.csv", delimiter=';')
    xpreproc1 = binaryencode(df)
    xpreproc2, x  = onehotencode(xpreproc1)

    newcase = xpreproc2[(xpreproc2['Case Number']==casenumber) & (xpreproc2['ItemInCase']==itemincasenumber)]
    customeroffer_newcase = tree.predict(newcase[featuresdt])
    probability = tree.predict_proba(newcase[featuresdt])
    return(newcase, customeroffer_newcase, probability)

a, b, c =  newcase_custoffer(1085988, 1)
```

## Optimization

There are many different avenues to optimize a classificatin tree. The two main areas to explore are the features themselves (volume of data and quality) and classification tree's parameters (max depth, training set size...).

Classification tree optimization is a vast topic and a lot can be said about it. Here let's just explore the impact of max depth and training set size on the average accuracy.

```python
# Decision Tree Optimisation Exploration Function
def dtopti(testsize, depth):
    
    # Predictor and response variables definition
    X = dfpp[featuresdt] # Predictor variables
    Y = dfpp['Customer Offer'] # Target variable
    
    """Rerun decision tree with different depth and save accuracy in a list"""
    # Split Data Training/Testing
    X_trainopti, X_testopti, y_trainopti, y_testopti = train_test_split(X, Y, test_size = testsize, random_state=42, stratify=Y)
    clfopti = DecisionTreeClassifier(criterion = "gini", random_state = 42, max_depth=depth) # Create Decision Tree
    clfopti = clfopti.fit(X_trainopti, y_trainopti) # Train Model
    y_predopti = clfopti.predict(X_testopti) # Predict response for X_test
    accuracyopti = clfopti.score(X_testopti, y_testopti)
    
    return accuracyopti

testsplit = []

for i in range(1,6,1):
    accuracy = []
    for j in range(1,30):
        accuracy.append(dtopti(float(i)/10, j))
    testsplit.append(accuracy)
```

```python
# Visualize accuracyopti
plt.figure(figsize=(20, 10))
for serie in range(len(testsplit)):
    plt.plot(testsplit[serie],label = 'Training Set Size %s%%'%((1-float(serie+1)/10)*100)) 
    
    
plt.legend()
plt.title('Impact of Training Set Size and Depth on Accuracy')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy (%)')
plt.show() 
```

<b>[Observations]</b><br>

- <b>Overfitting</b>: Increasing the training set size leads to a higher accuracy. true for the sample but not for the population, a high accuracy might also be linked to overfitting. Most litteratures recommand a training set size of 70 to 75% of sample size.
- <b>Pure leaves</b>: Past a certain depth the accuracy reaches an optimum, this is due to the nature of our features (categorical and discrete variables) and the relative small size of our the sample.

# Conclusion

Despite being at the end of the industry's value chain, after-sales support remains a powerful marketing vehicle. Performant after-sales support contributes to increasing brand loyalty and customer base.

Customer service representative roles are linked to very high turnover rates which makes it difficult for companies to ensure a uniform and seamless customer experience. 

This risk can be mitigated by controlling certain aspects of this role allowing the representatives to recenter the focus on strenghtening the customer relationship. Classification tree algorithms can be a solution to support representatives with warranty returns. The main advantages of CART algorithms are the low volume of data they need to be trained and their interpretability.

Python library Scikit-learn offers great options to implement such a solution. The algorithm produces a binary tree using a greedy approach. The concept of information gain is used to define each split (node). Gini Impurity and Entropy are the most common criteria used to calculate the information gain.
